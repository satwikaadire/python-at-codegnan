###List Methods
- item adding methods
 - append(item):it adds elements at the end of the string
 - extends(iterable):it adds the all iterator elements at end of the list
 - insert(pos,item):it adds the element at the specified position 

```python
lst=[10,20,30,40]
lst.append([100])
print(lst)
lst.extend([1,2,3])
print(lst)
lst.append([100,200])
print(lst)
lst.insert(0,1000)
print(lst)
```

###item removing methods
- remove(item):it removes the first occurance item from the list if it is exits in list othervise "ValueError"
- pop(pos):it removes and return the specefied position item if position is passed,otherwise it remove and returns the last item from the list
- clear():it removes all elements from the list


```python
lst=[10,20,30,20,40,50,20]
lst.remove(20)
print(lst)
```

```python
#pop
lst=[10,20,30,20,40,50,20]
val=lst.pop()
print(lst)
val2=lst.pop(1)
print(lst)
lst.clear()
print(lst)
```

```python
#index and icount()
#index(item): if the item is present inthe list it returns first occurance the index otherwise "ValueError"
#count(item):it returns the count of total occurances if item
lst=[10,20,30,20,40,50]
ind=lst.index(20)
print(ind)
print(lst.count(20))
```

```python
#sorting elements or items
#sort(reverse=false):by default it sorts the original list into ascending order,if reverse is true it sorts in decending order
lst=[1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
#charatersor words
lst=["a","d","s","c","d"]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```
