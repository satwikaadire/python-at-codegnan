### set characteristics


```python
# set is unordered collection
#set returns unique elements or it does not allow doplicates
s={1,2,3,2,3,"hi","a",2.02,(10,20,30)}
print(s)
#print(S[0]) #set does not support indexing

#list can't be used as set element
s={1,2,[3,2,3],"hi","a",2.02,(10,20,30)}
print(s)

```

### set Methods

```python
# element adding method
#add(ele): it adds the element to the set if element not in the set ,otherwise nothing
s={1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("HI")
print(s)
```

# element removing methods
- remove(item): it removes the item, if it exists in the set,other wise it raises "KeyError".
- discard(item): it is also used to remove elements,but it doesn't raise any keyerror 

```python
s={1,2,3,4,5}
s.add(100)
print(s)
s.remove(5)
print(s)
#s.remove(200) #raises key error
#print(s)
s.discard(4)
print(s)
s.discard(30)# it not raised an error
print(s)
```

```python
# set operations
#union(set):it returns all unique elements from all the sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
u=s1.union(s2,s3)
print(u)
res=s1|s2|s3
print(res)
```

```python
#intersection(set):it returns all the common elemsnts from the all the sets(symbol"&")
s1={1,3,5,7}
s2={4,5,6,7,3}
s3={1,2,8,9,}
intersection=s1.intersection(s2,s3)
print(intersection)
res=s1&s2&s3
print(res)
```

```python
#differences(set);it returns the a set elements those are not in another set
s1={1,3,5,7}
s2={4,5,6,7,}
s3={5,1,2,8,9}
intersection=s1.difference(s2,s3)
print(intersection)
res=s1-s2-s3
print(res)
print((s1|s2)-s3)
```

```python
#symmetric_difference(set);it returns the non commen elements from all the sets
s1={1,3,5,7}
s2={4,5,6,7,}
s3={5,1,2,8,9}
s_d=s1.symmetric_difference(s2)
print(s_d)
res=s1^s2^s3
print(res)
print((s1|s2)-s3)
```

```python
# issubset(set)
# issuperset(set)
#isdisjoint(set)
s1={1,2,3,3,4,5,9}
s2={2,3}
print(s2.issubset(s1))
print(s1.issuperset(s2))
print(s1.isdisjoint(s2))
print(dir(set))
```
