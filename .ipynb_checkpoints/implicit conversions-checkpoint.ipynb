```python
# integer to float,boolean,string,complex
num=10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
```

```python
#string to int,flaot,bool,list,tuple,set
#if string value is number
s="123456"
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#string to int,flaot,bool,list,tuple,set
#if string value is number
s="123456.12345678"
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#string to int,flaot,bool,list,tuple,set
#if string value is character
s="satwika"
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#list to string,tuple,set
lst=[10,20,30,40,20]
string=str(lst)
print(str(string),type(string))
print(tuple(lst))
print(set(lst))
```

```python
#tuple to string,list,set
tup=(12,34,5,6,33)
#print(lst(tup))
print(set(tup))
```

```python
#set to string,list,tuple
set={23,54,23,65,66,66}
print(set)
string=str(set)
print(str(string))
lst=list(set)
print(lst)
print(tuple(set))
```
