```python
# indexing
s="satwika"
print(s[4])
print(s[-5])
print(s[2:6:1])
```

```python
#slicing
#str_obj[start:end:step]
s="codegnan"
print(s[2:7])
print(s[-4:])
#it returns empty space it doest give error in slicing where as in indexing it will get error
print(s[5:2])
print(s[3:100])
print(s[::-1])
print(s[-1:-6:-1])
print(s[-2::-1])
print(s[-3::1])
print(s[:1])
```

```python
#empty string representation
str1=" "
str2=str()
print(type(str1))
print(type(str2))
```

```python
#string reading from user
#input() : this function is used to read dynamically from user and this function by default read it as string data type
s=input()
print(type(s))
print(s)
```

```python
#string operations
#1.string concatenation(adding)
#2.string repetation(*)
s1="code "
s2="gnan"
s3=s1+s2
print(s3)
print(s1+s1)
print(s1*5)
print(s1 * 7)
```

```python
#strings are immutable
s="codegnan"
res=s[:1]+'z'+s[2:]
print(res)
```
