##String method
- case conversion methods
   -upper():convert entire string into upper case.
   -lower():it converts entire string into lower case.
   -title():it converts the first letter of every word into upper and remining lower.
   -swapcase():it converts uppercase into lowercase and vice versa.
   -capitalize():it converts first character into uppercase and remining into lower case.

```python
string="The sky is blue and ocean is blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- Remove white space or pattern at ends of string
   - strip():by default it removes the white spaces at both the end of the string
   - lstrip():it removes the white space at left side of the string
   - rstrip():it removes the white space at right side of the string


```python
s="  satwika   "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@coe@@GJNJ@@@"
print(s)
print(s.strip("@"))
```

```python
#split(): it splits the entire string into list of substring based on specified separator
#by default it splits at space
string="The sky is blue and ocean is blue"
print(string.split())
print(string.split("is"))
print(string.split('e'))
```

```python
#'sep'.join(iterable):it joins the all iterable elements with separator
words=['the', 'sky', 'is', 'blue', 'and', 'ocean' 'is' 'bue']
print("".join(words))
print(" ".join(words))
print("@".join(words))
```

###substring index finding
#index(substring):if substring is exists in string, it returns the first occurencce of substring position,otherwise 'ValueError'
#find(substring):if substring is exists in string, it returns the first occurencce of substring position,otherwise '-1'

```python
string="The sky is blue and ocean is blue"
print(string.index('is'))
print(string.index('z'))
```

```python
string="The sky is blue and ocean is blue"
print(string.find('h'))
print(string.find('z'))
print(string.find('is',9))
print(string.find('e',10))
```

```python
#count(substring):it returns the total non-overlapping occurences of substring
s="ABCBCBCbbaabb"
print(s.count('B'))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count("b"))
print(s.upper())
print(s.upper().count("B"))
print(s.count("z"))
```

```python
#replace(old_string,new_string):it replace old string with the new string 
string="The sky is blue and ocean is blue"
print(string.replace("blue","Red"))
print(string.replace("e","EG"))
print(string.replace("e","EG",2))
```

###STRING CHECKING METHODS
-STRING CHECKING METHODS RETURNS THE BOOLEAN VALUE AS RESULT
-isupper():it return true,if entire string is in upper case
-islower():it return true,if entire string is in lower case
-isalpha():it return true,if entire string contains alphabets
-isdigit():it return true,if entire string contains digits
-isalnum():it return true,if entire string contains either alphabets or digits
-startswith(substring):it returns true,if string starts with substring
-endswith(substring):it returns true ,if string ends with substring.

```python
print("CODE".isupper())
print("CODE".islower())
print("code".islower())
print("123".isdigit())
print("123df".isdigit())
print("12ddc".isalnum())
print("4356".isalnum())
print("code".startswith('C'))
```
