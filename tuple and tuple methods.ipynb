```python
##tuple and tuples method
#ordered collection
t=(1,2,3,[10,20,30],23)
print(t)
```

```python
#if in tuple does not change length and position of the remaining elements then tuple is immutable
t=(1,2,3,[10,30,40],23)
t[3].append(100)
print(t)
```

```python
#empty tuple representation
#single element representation
t1=(10)
t2=(10,)
l1=[10]
t2=tuple()
print(type(t1),type(t2),type(l1))
```

```python
#tuple operations
#tuple cancatinaion
#tuple repetation
#tuple methods
#index()
#count()
t1=(10,22)
t2=(2,38)
print(t1+t2)
print(t1,t1)
print(t1*5)
```

```python
#index
t1=("hg",24,89,"gh",2.3,42,24,24,24)
print(t1.index(24))
print(t1.index("gh"))
print(t1.count(24))
```
