```python
#dictionary ordered collection
details={"name":"satwika","course":"python","batch":57,"age":21}
print(details)
```

```python
#dictionary values acces by using keys
details={"name":"satwika","course":"python","batch":57,"age":21}
print(details["name"])
#print(details["location"]) #key error

#updating the dictionary values
details["age"]=25
details["age"]=details["age"]+1
print(details["age"])

#adding a new items
details["location"]="Hyderabad"
print(details)
```

```python
#dictionary keys are unique
details={"age":21,"name":"satwika","age":28,"course":"python","batch":57,"age":29}
print(details)
```

```python
#list cannotbe used as dictionary
details={"age":21,"name":"satwika","age":28,"course":"python",[1,2,3]:57,"age":29,"marks":[24,45,5,7]}
print(details)
```

###dictionary Methods
- get(key,default_value:optional): it returns the value if key is exists in dictionary,otherwise it returns default value.

```python
details={"age":21,"name":"satwika","age":28,"course":"python","batch":57, "age":29}
print(details["name"])
print(details.get("name"))
print(details["marks"]) #key error because in above dictionary marks is not available where as this get it returns vdefault value.
print(details.get("marks","not there"))
```

- updata(new_dictionary): THis method updates the values if new _dictionary keys already exists in dictionary,otherwise it adds new items to dictionary
- 

```python
details={"name":"satwika","course":"python","batch":57,"age":21}
details.update({"batch":67,"course":"java","Location":"hyderabad"})
print(details)
```

- 'keys()' :this method returns the list dictionary keys
- 'values()' :this method returns the list dictionary values
- 'items()' : this method return the list of tuple of key value pairs

```python
details={"age":21,"name":"satwika","age":28,"course":"python","batch":57,"age":29}
print(details.keys())
print(details.values())
print(details.items())
```

- popitem(): it return and removes the item as tuple
- pop(key): this method removes the specified key item and return the specified key's value,otherwise keyvalue

```python
details={"age":21,"name":"satwika","age":28,"course":"python","batch":57,}
item=details.popitem()
print(item)
print(details)
val=details.pop("name")
print(val)
print(details)
```
