### Input Formating

```python
#reading integer from user
num=int(input("enter an number:"))
print(type(num))
print(num)
```

```python
#taking float values from the user
val=float(input())
print(type(val))
print(val)
```

```python
#taking complex values fro the user
val=complex(int(input()),int(input()))
print(type(val))
print(val)
```

```python
#taking boolean values from the user
val=bool(int(input()))
print(type(val))
print(val)
```

```python
#taking string values from the user
num=str(input())
print(type(num))
print(num)
```

```python

s=["1","s"]
print(type(s[0]))
```

```python
#reading list of integers from user
lst=list(map(int,input().split()))
print(lst)
print(type(lst[2]))

```

```python
#reading string from the user
lst=list(map(str,input().split(",")))
print(lst)
print(type(lst[3]))
```

```python
#reading tuple of integers from user
lst=list(map(tuple,input().split()))
print(lst)
```

```python
#reading set of integers from user
lst=list(map(set,input().split()))
print(lst)
```

```python
#eval(): this function evaluates the string type functions
'''print("1+89")
print(eval("1+89"))'''
#taking input from user
val=eval(input())
print(type(val))
print(val)
```

```python
#access specifiers
''''string - %s
integer - %d
float - %f'''
name,age="satwika",20
print("my name is call %s and iam %d years old"%(name,age))
```

```python
#3 dot format
name,age="satwika",20
print("my name is call {} and iam {} years old".format(name,age))
```

```python
#f format
name,age="satwika",20
print(f"my name is call {name} and iam {age} years old")
print(f"my name is call {age} and iam {name} years old")
```
